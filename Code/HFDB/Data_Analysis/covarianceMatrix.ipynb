```python
import matplotlib
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
cols = ['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine',
        'serum_sodium','sex','smoking','time','DEATH_EVENT']

df = pd.read_csv('../../../HeartFailure_Database/heart_failure_clinical_records_dataset.csv')
print(df.shape)
```

```python
removedColumns = cols[1:]
covMatrix = pd.DataFrame.cov(df[removedColumns])
print(covMatrix)
covMatrix = np.power(covMatrix, 2)
covMatrix = np.power(covMatrix, 1/20)
print(covMatrix)
sn.heatmap(covMatrix)
plt.rcParams["figure.figsize"] = (30, 24)
plt.title("Covariance Matrix of Breast Cancer Features Scaled to x^1/20", fontdict={'fontsize': 29})
plt.show()
```
