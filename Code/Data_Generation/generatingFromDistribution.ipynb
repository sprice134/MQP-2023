```python
import matplotlib
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
cols = ['ID', 'Diagnosis', 'radius', 'texture', 'perimeter', 'area', 'smoothness', 'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension',
        'radius_SE', 'texture_SE', 'perimeter_SE', 'area_SE', 'smoothness_SE', 'compactness_SE', 'concavity_SE', 'concave_points_SE', 'symmetry_SE', 'fractal_dimension_SE',
        'radius_MAX', 'texture_MAX', 'perimeter_MAX', 'area_MAX', 'smoothness_MAX', 'compactness_MAX', 'concavity_MAX', 'concave_points_MAX', 'symmetry_MAX', 'fractal_dimension_MAX']
df = pd.read_csv('../../Wisconsin_Database/wdbc.data', header=None)
df.columns = cols
df['Diagnosis'] = df['Diagnosis'].map(
                   {'M':1 ,'B':0})
print(df.shape)
```

```python
df = df[cols[1:]]
for i in df:
    print('Column: {}'.format(i))
    print('\tMin: {}, Max: {}, Mean: {:.02f}, Std Dev: {:.02f}'.format(min(df[i]), max(df[i]), sum(df[i])/len(df[i]), np.std(df[i])))
    hist, bins = np.histogram(df[i], bins=20)
    plt.hist(df[i], bins = bins)
    plt.show()
    print('New Graph')

    bin_midpoints = bins[:-1] + np.diff(bins)/2
    cdf = np.cumsum(hist)
    cdf = cdf / cdf[-1]
    values = np.random.rand(150000)
    value_bins = np.searchsorted(cdf, values)
    random_from_cdf = bin_midpoints[value_bins]
    plt.hist(random_from_cdf, 20, color = 'purple')
    plt.show()

    '''(n, bins, patches) = plt.hist(df[i], bins=20)
    print(n)
    print(bins)
    plt.show()'''
```
